Общий синтаксис get запросов в REST API *Materials Project*:

`https://www.materialsproject.org/rest/v2/{request_type}[/{identifier}][/{parameters}]`

Первая часть запроса (URI) `https://www.materialsproject.org/rest/v2/`, содержит адрес REST запроса. 

- `v2` обозначет 2-ую версию MAPI, эта часть необходима чтобы сделать запросы гибкими в импелементациях в ПО при будущих обновлениях API.

- `{request_type}` определяет информационный раздел или операцию запроса. Поддерживаемые типы запросов: `materials`, `battery`, `reaction`, `mpquery` и `api_check`.

- `{identifier}` идентификатор для специализированных запросов. Зависит от типа запроса `{request_type}`, не является обязательным полем.

- `{parameters}` - запросы дополнительной информации

*Подробнее*: https://docs.materialsproject.org/open-apis/the-materials-api/#ssl-encryption

Одним из наиболее важных дескрипторов стабильности материалов является параметр $E_{hull}$.

Методика его определения включает перебор всех возможных комбинаций соединений, которые состоят из тех же атомов, что и рассматриваемый материал и в сумме дают те же стехиометрические коэффициенты.
Например, структура $Co_2O_3$ будет проверяться на разложение по сравнению с другими структурами $Co_2O_3$, со смесями $Co$ и $O_2$ и со смесями $CoO$ и $O_2$ (учитывая все степени окисления элементов).

***Если сумма энергий продуктов разложения меньше энергии исследуемого материала, то такое состояние энергетически более выгодно для системы и такой материал - нестабилен***

Положительная $E_{hull}$ указывает на то, что этот материал нестабилен по отношению к разложению.

$E_{hull} = 0$ указывает на то, что это наиболее стабильный материал по своему составу.

Таким образом $E_{hull}$ - это расстояние до выпуклой оболочки потенциальной энергии или, другими словами - разница энергии рассматриваемого материала и такого сочетания веществ, суммарная энергия которого минимальна при заданных стехиометрических соотношениях.

Важно, что эта оболочка представленна только некоторыми точками, т.к. не каждое стехиометрическое соотношение может давать стабильный состав, так же как и не в каждой композиционной точке (~ вертикальной линии) могут вообще существовать вещества.

Для однопоточной обработки запросов потребуется функция `get_data()` (в целом, MAPI поддерживает асинхронные запросы):

```python
def get_data(formula):
    prefix = ('https://www.materialsproject.org/rest/v2/materials/'+formula+'/vasp/')
    request_type = 'materials'
    request = rq.get(prefix, headers = header)
    response = request.json()
    data = response['response']
    return data

data_array = []

for f in formulas:
    data = get_data(f)
    
    #Select all stable phases from initial dataset
    for item in data:
        if item['e_above_hull'] == 0:
            data_array.append(item)
```

Воспользуемся методикой VRH* для пересчета уже имеющихся в БД упругих характеристик и, затем, сравним их с референсными данными:

https://wiki.materialsproject.org/Elasticity_calculations

```python
def get_elasticity(elasticity):
    elastic_tensor = elasticity['elastic_tensor']
    c11, c12, c13 = elastic_tensor[0][0], elastic_tensor[0][1], elastic_tensor[0][2]
    c21, c22, c23 = elastic_tensor[1][0], elastic_tensor[1][1], elastic_tensor[1][2]
    c31, c32, c33 = elastic_tensor[2][0], elastic_tensor[2][1], elastic_tensor[2][2]
    c44, c55, c66 = elastic_tensor[3][3], elastic_tensor[4][4], elastic_tensor[5][5]

    compliance_tensor = np.linalg.pinv(elastic_tensor) #only for 3d range tensor (genereally cubic lattices)
    s11, s12, s13 = compliance_tensor[0][0], compliance_tensor[0][1], compliance_tensor[0][2]
    s21, s22, s23 = compliance_tensor[1][0], compliance_tensor[1][1], compliance_tensor[1][2]
    s31, s32, s33 = compliance_tensor[2][0], compliance_tensor[2][1], compliance_tensor[2][2]
    s44, s55, s66 = compliance_tensor[3][3], compliance_tensor[4][4], compliance_tensor[5][5]

    # bulk modulus calculation according Voigt-Reuss-Hill
    kv = ((c11 + c22 + c33) + 2*(c12 + c23 + c31)) / 9
    kr = 1 / ((s11 + s22 + s33) + 2*(s12 + s23 + s31))
    kvrh = (kv + kr) / 2 # all three are in GPa
    
    # shear modulus calculation according Voigt-Reuss-Hill
    gv = ((c11 + c22 + c33) - (c32 + c23 + c31) + 3*(c44 + c55 + c66)) / 15
    gr = 15 / (4*(s11 + s22 + s33) - 4*(s12 + s23 + s31)+3*(s44 + s55 + s66))
    gvrh = (gv + gr) / 2 # all three are in GPa
        
    # calculated shear and bulk moduli
    shear_modulus = round(gvrh, 0)
    bulk_modulus = round(kvrh, 0)
        
    return c11, c12, c44, shear_modulus, bulk_modulus
```

Обратите внимание, что тензор податливости $S_{ij}$ в общем случае рассчитывается иначе: https://ru.wikipedia.org/wiki/Нотация_Фойгта

Приведенная формула справделива только для кубических решеток и тензоров констант упругости третьего порядка.

Добавим рассчитанные нами данные в датафрейм:

```python
# add obtained values to DataFrame
bulk_array = []
shear_array = []

for item in df['elasticity']:
    elastic_properties = get_elasticity(item)
    shear_array.append(elastic_properties[3]) 
    bulk_array.append(elastic_properties[4])
        
df['Calculated_Shear'] = shear_array
df['Calculated_Bulk'] = bulk_array
df
```

Поскольку видно, что методика применена верно, теперь можно получить новую !практически-важную! характеристику упругости материалов - **модуль Юнга** (E, ГПа) (https://ru.wikipedia.org/wiki/Модуль_Юнга)

$$E = \frac{K_{VRH} \cdot G_{VRH}}{3 K_{VRH} + G_{VRH}}$$

Для этого модифицируем имеющуюся функцию:

```python
def get_elasticity(elasticity):
    elastic_tensor = elasticity['elastic_tensor']
    c11, c12, c13 = elastic_tensor[0][0], elastic_tensor[0][1], elastic_tensor[0][2]
    c21, c22, c23 = elastic_tensor[1][0], elastic_tensor[1][1], elastic_tensor[1][2]
    c31, c32, c33 = elastic_tensor[2][0], elastic_tensor[2][1], elastic_tensor[2][2]
    c44, c55, c66 = elastic_tensor[3][3], elastic_tensor[4][4], elastic_tensor[5][5]

    compliance_tensor = np.linalg.pinv(elastic_tensor) #only for 3d range tensor (genereally cubic lattices)
    s11, s12, s13 = compliance_tensor[0][0], compliance_tensor[0][1], compliance_tensor[0][2]
    s21, s22, s23 = compliance_tensor[1][0], compliance_tensor[1][1], compliance_tensor[1][2]
    s31, s32, s33 = compliance_tensor[2][0], compliance_tensor[2][1], compliance_tensor[2][2]
    s44, s55, s66 = compliance_tensor[3][3], compliance_tensor[4][4], compliance_tensor[5][5]

    # bulk modulus calculation according Voigt-Reuss-Hill
    kv = ((c11 + c22 + c33) + 2*(c12 + c23 + c31)) / 9
    kr = 1 / ((s11 + s22 + s33) + 2*(s12 + s23 + s31))
    kvrh = (kv + kr) / 2 # all three are in GPa
    
    # shear modulus calculation according Voigt-Reuss-Hill
    gv = ((c11 + c22 + c33) - (c32 + c23 + c31) + 3*(c44 + c55 + c66)) / 15
    gr = 15 / (4*(s11 + s22 + s33) - 4*(s12 + s23 + s31)+3*(s44 + s55 + s66))
    gvrh = (gv + gr) / 2 # all three are in GPa
        
    # calculated shear and bulk moduli
    shear_modulus = round(gvrh, 0)
    bulk_modulus = round(kvrh, 0)
        
    # Young modulus
    E = 9 * (kvrh*gvrh)/(3*kvrh + gvrh)
    young_modulus = round(E, 2)
    
    return c11, c12, c44, shear_modulus, bulk_modulus, young_modulus
```

```python
young_moduli = []

for item in df['elasticity']:
    elastic_properties = get_elasticity(item)
    young_moduli.append(elastic_properties[5])

df['Young'] = young_moduli
df
```

Добавим новый критерий, который не содержит расчетных данных, и может быть получен только из данных о составе вещества.

**Концентрация валентных электронов (VEC)** - это количество валентных электронов, приходящееся на один относительный атом вещества.

$VEC = \frac{\sum (n*Val)}{N}$

где $n$ - число атомов каждого сорта;
    $Val$ - число валентных электронов, учтенных в расчете, для данного сорта атомов;
    $N$ - общее количество атомов или сумма стехиометрических коэффициентов

Во всех рачсетах MP используются чаще всего рекомендованные потенциалы, поэтому воспользуемся таблицей рекомендованных для использования в VASP потенциалов.

```python
valency = pd.read_csv(filepath_or_buffer = "Default_PAW_potentials_VASP.csv", sep = ';')
valency.sample(3)
```

Добавим значения $VEC$ для каждой записи в датафрейме с помощью функции `get_VEC`:

```python
formulas_array = df.reset_index()
formulas_array = formulas_array['unit_cell_formula']

#VEC calculation function
def get_VEC(dict):
    S = sum(dict.values())
    valency_sum = 0
    for elem in dict.keys():
        V = valency[valency['Element']  == elem].valency.values[0]
        N = dict.get(elem)
        valency_sum += V*N
    VEC = valency_sum/S
    
    return VEC

VEC_array = []
for formula in formulas_array:
    VEC_array.append(get_VEC(formula))

df['VEC'] = VEC_array
df.sample(3)
```

Посмотрим на матрицу корреляций снова:

```python
corr_matrix = df.corr()

#exclude constant and empty data
corr_matrix = corr_matrix.drop(['nelements', 'e_above_hull', 'spacegroup_number'], axis='columns')
corr_matrix = corr_matrix.dropna()

plt.figure(figsize=(6,6))
plt.imshow(corr_matrix, cmap='coolwarm')
plt.colorbar()  # add color intensity map

plt.xticks(rotation=90)

plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.yticks(range(len(corr_matrix)), corr_matrix.index)
```
